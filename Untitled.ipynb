```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score

# Load the training data
train_data = np.loadtxt('train.txt')
x_train = train_data[:, 0]  # Temperature
y_train = train_data[:, 1]  # Ice cream sales

# Load the test data
test_data = np.loadtxt('test.txt')
x_test = test_data[:, 0]
y_test = test_data[:, 1]

# Plot the original training data
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', s=10, label='Training Data')
plt.xlabel('Temperature')
plt.ylabel('Number of Ice Creams Sold')
plt.title('Ice Cream Sales vs Temperature')
plt.legend()
plt.show()

# Fit a linear regression model on the training data
model = LinearRegression()
model.fit(x_train.reshape(-1, 1), y_train)
beta0_hat = model.intercept_
beta1_hat = model.coef_[0]

# Print the linear regression coefficients
print("Linear Model Coefficients:")
print(" (Intercept):", beta0_hat)
print(" (Slope):", beta1_hat)

# Predict on both training and test sets
y_train_pred = model.predict(x_train.reshape(-1, 1))
y_test_pred = model.predict(x_test.reshape(-1, 1))

# Handle potential zero values in x_train for log transformation
epsilon = 1e-8  # Small constant to avoid log(0)
features = {
    'cos_x': np.cos(x_train),
    'log_x': np.log(x_train + epsilon),
    'cos_4x': np.cos(4 * x_train),
    'sin_3x': np.sin(3 * x_train),
    'sin_5x': np.sin(5 * x_train),
    'sin2x_cos2x': np.sin(2 * x_train) * np.cos(2 * x_train)
}

# Convert features to a single matrix X for training data
X_train = np.column_stack([x_train] + list(features.values()))

# Fit Lasso and Ridge regression models on training data with non-linear features
lasso = Lasso()
lasso.fit(X_train, y_train)
ridge = Ridge()
ridge.fit(X_train, y_train)

# Print the coefficients from Lasso and Ridge regression
print("\nLasso Regression Coefficients:", lasso.coef_)
print("Ridge Regression Coefficients:", ridge.coef_)

# Model evaluation on test data
print("\nEvaluation on Test Data:")
mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)
print("Mean Squared Error (MSE):", mse)
print("R Score:", r2)

# Plot training and test predictions to visualize overfitting
plt.figure(figsize=(10, 6))

# Plot training data with predictions
plt.scatter(x_train, y_train, color='blue', s=10, label='Training Data')
plt.plot(x_train, y_train_pred, color='darkblue', linewidth=2, label='Training Prediction')

# Plot test data with predictions
plt.scatter(x_test, y_test, color='orange', s=10, label='Test Data')
plt.plot(x_test, y_test_pred, color='red', linestyle='--', linewidth=2, label='Test Prediction')

# Set chart details
plt.xlabel('Temperature')
plt.ylabel('Number of Ice Creams Sold')
plt.title('Model Overfitting Visualization: Training vs Test Predictions')
plt.legend()
plt.grid(True)
plt.show()
```
